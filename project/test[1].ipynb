```python
# companyswholedata = 
nse_ticker = ["NAVA", "BLUEJET","KFINTECH"]
```

```python
import pandas as pd
import yfinance as yf

for companies in nse_ticker:
    companyswholedata = pd.read_excel('Stock_data.xlsx')
    print(len(companyswholedata))
    
    companyswholedata_rowwise["tickers"]= companies
    for infos in companysinfo:
        try:
            companies_stock_data = yf.Ticker(f"{companies}.NS")
            if companies_stock_data.info:
                companyswholedata_rowwise[infos] = companies_stock_data.info.get(infos, None)
        except Exception as e:
           (f"Error retrieving data for {companies}: {e}")

    for infos in companyscashflow:
        try:
            companies_stock_data = yf.Ticker(f"{companies}.NS")
            if companies_stock_data.cashflow is not None and infos in companies_stock_data.cashflow.index:
                companyswholedata_rowwise[infos] = companies_stock_data.cashflow.loc[infos].values.tolist()
        except Exception as e:
           (f"Error retrieving cashflow data for {companies}: {e}")

    for infos in companysbalancesheet:
        try:
            companies_stock_data = yf.Ticker(f"{companies}.NS")
            if companies_stock_data.balancesheet is not None and infos in companies_stock_data.balancesheet.index:
                companyswholedata_rowwise[infos] = companies_stock_data.balancesheet.loc[infos].values.tolist()
        except Exception as e:
            (f"Error retrieving balancesheet data for {companies}: {e}")

    for infos in companysfinancialsinfo:
        try:
            companies_stock_data = yf.Ticker(f"{companies}.NS")
            if companies_stock_data.financials is not None and infos in companies_stock_data.financials.index:
                companyswholedata_rowwise[infos] = companies_stock_data.financials.loc[infos].values.tolist()
        except Exception as e:
           (f"Error retrieving financials data for {companies}: {e}")

    # Use pd.concat() instead of append()
    companyswholedata = pd.concat([companyswholedata, pd.DataFrame([companyswholedata_rowwise])], ignore_index=True)

    # Save to Excel
    companyswholedata.to_excel('Stock_data.xlsx', index=False)

    # Reset the row data
    companyswholedata_rowwise = companyswholedata_rowwise_copy


```
